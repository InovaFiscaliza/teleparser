# Número de Blocos

```python
parquet = pasta.ls().filter(lambda x: x.suffix == '.parquet')
parquet
```

```python
df = pd.read_parquet(parquet[0]).reset_index(drop=True).astype('category')
df.info()
```

```python
df['error'].value_counts()
```

```python
df.sample(10)
```

* Extrair identificador

```python
df['Unique-Id'] = df['Session-Id'].str.split(';', n=1, expand=True)[0]
df['Unique-Id'].value_counts()
```

* Vamos criar uma coluna `Identifier` para diferenciar os "tipos" de CDR de acordo dentro do mesmo arquivo
> A HUAWEI é a única que aparentemente registra a coluna `Vendor-ID` como dado principal em primeiro nível. Então todos os registros não nulos dessa coluna podem ser identificados como HUAWEI.

```python
df['Vendor-Id'].notna().sum().item()
```

```python
df.loc[df['Vendor-Id'].notna(), 'Identifier'] = "HUAWEI"
```

Pelos dados de exemplo traduzidos repassados pela Claro, todos os registros `HUAWEI` são do tipo `TAS`. Vamos criar outra coluna com nome `Type` que será util para diferenciar os demais dados.

```python
df.loc[df["Vendor-Id"].notna(), "Type"] = "TAS"
```

Nesse caso, os campos não nulos `AVPVendor-ID` da `HUAWEI` estão utilizando o código `2011` que identifica "inequivocamente" a `HUAWEI` como operadora, no entanto com esse campo está presente em poucos registros, não podemos utilizá-lo como identificador.

```python
df.loc[df["Unique-Id"].str.startswith("pcscf"), "Type"] = "SBG"
df.loc[df["Unique-Id"].str.startswith("pcscf"), "Identifier"] = "ERICSSON"
```

```python
df.loc[df["Unique-Id"].str.startswith("scscf"), "Type"] = "IMS"
df.loc[df["Unique-Id"].str.startswith("scscf"), "Identifier"] = "ERICSSON"
```

Existem tanto registros da `ERICSSON` quanto `HUAWEI` cujo `Unique-Id` começam com `tas`, no entanto como identificamos a `HUAWEI` pela presença da coluna `Vendor-Id` podemos simplesmente identificar os demais como `ERICSSON`:

```python
slice = df["Unique-Id"].str.startswith("tas") & df["Vendor-Id"].isna()
df.loc[slice, "Type"] = "TAS"
df.loc[slice, "Identifier"] = "ERICSSON"
```
